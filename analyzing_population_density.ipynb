```python
import pandas as pd
import geopandas as gpd
import folium
import branca.colormap as cm
from folium.features import GeoJsonTooltip
```

```python
path_csv_data = 'usa_population_2019.csv'
population_data = pd.read_csv(path_csv_data)
```

```python
population_data.head(10)
```

```python
population_json_data = 'us-states.json'
gdf = gpd.read_file(population_json_data)
```

```python
gdf.head()
```

```python
gdf.crs
```

```python
mereged_data = gdf.merge(population_data, left_on='name', right_on='Geographic Area')
mereged_data.head()
```

```python
mereged_data['area_km2'] = mereged_data['geometry'].to_crs(3395).area / 10**6
mereged_data['population_density'] = mereged_data['Total Resident Population'] / mereged_data['area_km2']
mereged_data
```
